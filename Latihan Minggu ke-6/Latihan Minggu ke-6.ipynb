```python
import pandas as pd
import numpy as np
data=pd.read_csv("D:\weatherAUS.csv")
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
data['MinTemp'] = data['MinTemp'].fillna(12.194034)
data['MaxTemp'] = data['MaxTemp'].fillna(23.221348)
data['Rainfall'] = data['Rainfall'].fillna(2.360918)
data['Evaporation'] = data['Evaporation'].fillna(5.468232)
data['Sunshine'] = data['Sunshine'].fillna(7.611178)
data['WindGustSpeed'] = data['WindGustSpeed'].fillna(40.035230)
data['WindSpeed9am'] = data['WindSpeed9am'].fillna(14.043426)
data['WindSpeed3pm'] = data['WindSpeed3pm'].fillna(18.662657)
data['Humidity9am'] = data['Humidity9am'].fillna(68.880831)
data['Humidity3pm'] = data['Humidity3pm'].fillna(51.539116)
data['Pressure9am'] = data['Pressure9am'].fillna(1017.64994)
data['Pressure3pm'] = data['Pressure3pm'].fillna(1015.255889)
data['Cloud9am'] = data['Cloud9am'].fillna(4.447461)
data['Cloud3pm'] = data['Cloud3pm'].fillna(4.509930)
data['Temp9am'] = data['Temp9am'].fillna(16.990631)
data['Temp3pm'] = data['Temp3pm'].fillna(21.68339)
data['RISK_MM'] = data['RISK_MM'].fillna(2.360682)



np.sum(data.isnull())
```

```python
from scipy.stats import mode

data['WindGustDir'] = data['WindGustDir'].fillna(data['WindGustDir']).dropna().mode()[0]
data['WindDir9am'] = data['WindDir9am'].fillna(data['WindDir9am']).dropna().mode()[0]
data['WindDir3pm'] = data['WindDir3pm'].fillna(data['WindDir3pm']).dropna().mode()[0]
data['RainToday'] = data['RainToday'].fillna(data['RainToday']).dropna().mode()[0]
data['RainTomorrow'] = data['RainTomorrow'].fillna(data['RainTomorrow']).dropna().mode()[0]
np.sum(data.isnull())
```

```python
from scipy import stats
z = np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data = data[(z < 3).all(axis=1)]
print(data.shape)
```

```python
df=data.drop(columns=['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'],axis=1)
df.shape
```

```python
x=df.iloc[0:134753,1:17]
x.tail()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm l2, other are l1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
#3sigmoid
xsigmoidal=((1-2.718281828**(-xnorm))/(1+2.718281828**(-xnorm)))
xsigmoidal=pd.DataFrame(xsigmoidal)
xsigmoidal.tail()
```

```python
#decimal scalling
xdecimal=x/(10**3) #rumus asli x/(10**i) , i dapat ditentukan dg bebas
xdecimal=pd.DataFrame(xdecimal)
xdecimal.tail()
```

```python
#softmax
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```
