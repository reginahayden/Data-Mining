```python
import pandas as pd
import numpy as np
```

```python
#import dataset
pcademo=pd.read_csv("D:\pcademo.csv")
x=pcademo.iloc[0:51,1:20]
x.tail()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') #default norm l2, other are l1 and max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
#minmax
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-5,1))
scaler.fit(x) #rumus bawaan packcage Minmaxscaler
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
#3sigmoid
xsigmoidal=((1-2.718281828**(-xnorm))/(1+2.718281828**(-xnorm)))
xsigmoidal=pd.DataFrame(xsigmoidal)
xsigmoidal.tail()
```

```python
#decimal scalling
xdecimal=x/(10**3) #rumus asli x/(10**i) , i dapat ditentukan dg bebas
xdecimal=pd.DataFrame(xdecimal)
xdecimal.tail()
```

```python
#softmax
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```
