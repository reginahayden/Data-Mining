```python
import pandas as pd 
import numpy as np
data1 = pd.read_csv("D:\wikipedia.csv" , sep=";")
data1.head()
np.sum(data1.isnull())
```

```python
data1.describe()
```

```python
data1['YEARSEXP'] = data1['YEARSEXP'].fillna(12.791667)
np.sum(data1.isnull())
```

```python
data1.to_csv("D:/wikipedia2.csv")
```

```python
from scipy import stats
z = np.abs(stats.zscore(data1._get_numeric_data()))
print(z)
data1 = data1[(z < 3).all(axis=1)]
print(data1.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
x=data1.iloc[0:160,0:50]
scaler=MinMaxScaler(feature_range=(-5,1))
scaler.fit(x) 
x_scaled=scaler.transform(x)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
x_scaled.to_csv("D:/wikipedia3.csv")
```
