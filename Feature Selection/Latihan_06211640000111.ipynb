# Principal Component Analysis (PCA)

```python
#principal component analysis
import numpy as np
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
%matplotlib inline
```

```python
pcademo = pd.read_csv("D:/pcademo.csv")
pcademo.head()
```

```python
#set up the feature
feature = pcademo.iloc[0:51,1:20]
feature.tail()
```

```python
#convert it to numpy arrays
x=feature.values

#scaling the value
x=scale(x)
pca=PCA(n_components=19)
pca.fit(x)

#the amount of variance that each PC explain
var=pca.explained_variance_ratio_

#cumulative variance explain
var1=np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)

print (var1)
plt.plot(var1)
               
```

```python
# looking at above plot (i'm taking 30 variables)
pca=PCA(n_components=4)
pca.fit(x)
x1=pca.fit_transform(x)
print(x1)
```

```python
feature=pd.DataFrame(x1) #mengubah array ke dataframe agar bisa diolah lebih lanjut
feature.head()
```
