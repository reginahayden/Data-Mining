## Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("D:\Data Mining\movie-metadata.csv")
data.head()
```

```python
#descriptive statistics
data.describe()
```

```python
#number of NA for each variable
#untuk melihat jumlah data yg kosong
np.sum(data.isnull())
```

```python
#jika data yg kosong kurang dari 10% jumlah data maka diabaikan, jika lebih maka : 
data['budget'] = data['budget'].fillna(3.975262e+07) #mengisi nilai yang kosong dg rata" pd variabel budget
np.sum(data.isnull())
```

```python
data['gross'] = data['gross'].fillna(4.846841e+07) #mengisi nilai yang kosong dg rata" pd variabel gross
np.sum(data.isnull())
```

```python
# missing value
# jika per variabel krg dr 10% diabaikan/diimputasi, tetap dipakai
# kalau diatas 15% masuk kandidat variabel dihapus
# kalau sudah lebih dari 50% variabel tsb dibuang dari pengamatan
```

```python
data['duration'] = data['duration'].fillna(107.201074) #mengisi nilai yang kosong dg rata" pd variabel 
data['facenumber_in_poster'] = data['facenumber_in_poster'].fillna(1.371446) #atau bisa dengan input fillna.((data['facenumber_in_poster'].mean())) tanpa hrs mencari
data['num_user_for_reviews'] = data['num_user_for_reviews'].fillna(272.770808)
data['num_critic_for_reviews'] = data['num_critic_for_reviews'].fillna(140.194272)
data['director_facebook_likes'] = data['director_facebook_likes'].fillna(686.621709)
data['actor_1_facebook_likes'] = data['actor_1_facebook_likes'].fillna(6561.323932)
data['actor_2_facebook_likes'] = data['actor_2_facebook_likes'].fillna(1652.080533)
data['actor_3_facebook_likes'] = data['actor_3_facebook_likes'].fillna(645.009761)
np.sum(data.isnull())
```

```python
# fill missing value with mode for qualitative/categorical data 
from scipy.stats import mode
data['country'] = data['country'].fillna(data['country']).dropna().mode()[0]
```

```python
data.isnull().sum()
```

```python
#mengatasi outlier untuk univariate
from scipy import stats
z = np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data = data[(z < 3).all(axis=1)]
print(data.shape)
```
